```python
from PIL import Image, ImageDraw
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#img = cv2.imread('exemplo.jpg')
img = Image.open('exemplo.jpg').convert('RGB')
imgplot = plt.imshow(img)
plt.show()
```

```python
rotated = img.rotate(-11)
imgplot = plt.imshow(rotated)
plt.show()
```

```python
shadow = Image.new('RGB', rotated.size, color='black')
rotated.paste(shadow, (300, 3800), mask=shadow)

mask = Image.new("L", rotated.size, 0)
draw = ImageDraw.Draw(mask)
draw.polygon([(0, 1600), (0, 1400), (3800, 1400), (3800, 1600)], fill=255)
im = Image.composite(rotated, shadow, mask)
im = np.array(im)[1400:1600, 0:3800]
im = im[:, :, ::-1].copy() 
# rotatedplot = plt.imshow(rotated.rotate(-10)[, 1400:1600])
imgplot = plt.imshow(im)
plt.show()
```

```python
cv2.imwrite('out.jpg', im)
```
